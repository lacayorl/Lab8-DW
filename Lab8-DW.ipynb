#### Imputación por moda

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')
imp_mode = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
titanic['new_sibsp'] = imp_mode.fit_transform(titanic[['SibSp']])
imp_mode.fit_transform(titanic[['SibSp']])
```

```python
titanic['new_parch'] = imp_mode.fit_transform(titanic[['Parch']])
imp_mode.fit_transform(titanic[['Parch']])
```

```python
titanic['new_embarked'] = imp_mode.fit_transform(titanic[['Embarked']])
imp_mode.fit_transform(titanic[['Embarked']])
```

#### Imputación por mediana

```python
imp_median.fit_transform(titanic[['Age']])
```

```python
titanic['new_age'] = imp_median.fit_transform(titanic[['Age']])
```

```python
titanic['new_fare'] = imp_median.fit_transform(titanic[['Fare']])
imp_median.fit_transform(titanic[['Fare']])
```

### Normlice las columnas numéricas por los métodos:

* Standarization
* MinMaxScaling
* MaxAbsScaler

#### Standarization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
titanic_z = titanic.copy()
for col in titanic_z.select_dtypes(include=['float', 'int']).columns:
    titanic_z[col+'_z'] = scaler.fit_transform(titanic_z[[col]])
```

```python
titanic_z.head()
```

```python
titanic_z.describe()
```

#### MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
titanic_norm = titanic.copy()
for col in titanic_norm.select_dtypes(include=['float', 'int']).columns:
    titanic_norm[col+'_norm'] = scaler.fit_transform(titanic_norm[[col]])
```

```python
titanic_norm.head()
```

```python
titanic_norm.describe()
```
